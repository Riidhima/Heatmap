```python
m = folium.Map(location=[40.7128,-74.0060],
                        zoom_start=11,
                        tiles="CartoDB dark_matter")

for index, row in crimes_hm.iterrows():
    folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=3,
                        popup=row['LAW_CAT_CD'],
                        fill_color="#3db7e4", 
                       ).add_to(m)
    
```
